```python
#data loading
file_path='Resources/shopping_data.csv'
df_shopping = pd.read_csv(file_path, encoding='ISO-8859-1')
df_shopping.head(5)
```

```python
#columns
df_shopping.columns
```

```python
#list dataframe data types
df_shopping.dtypes
```

```python
#find null values
for column in df_shopping.columns:
    print(f'Column {column} has {df_shopping[column].isnull().sum()} null values')
    
    
```

```python
#drop null rows
df_shopping = df_shopping.dropna()
```

```python
#find duplicate entries
print(f'Duplicate entries: {df_shopping.duplicated().sum()}')
```

```python
#remove the customer ID column
df_shopping.drop(columns=['CustomerID'], inplace=True)
df_shopping.head()
```

```python
#transform string column
def change_string(member):
    if member =='Yes':
        return 1
    else:
        return 0
    
df_shopping['Card Member']=df_shopping['Card Member'].apply(change_string)
df_shopping.head()
```

```python
#transfrom annual income
df_shopping['Annual Income'] = df_shopping['Annual Income']/1000
df_shopping.head()
```

```python
df_shopping_renamed = df_shopping.rename (columns = {'Card Member':'CardMember','Age':'Age','Annual Income':'AnnualIncome','Spending Score (1-100)':'SpendingScore'} )
df_shopping_renamed
```

```python
#saving cleaned data
file_path = '/Users/paolaescamillamartinez/Desktop/Classwork/Cryptocurrencies/Resources/shopping_data_cleaned.csv'
df_shopping_renamed.to_csv(file_path, index=False)
```
